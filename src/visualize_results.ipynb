# Visualize Results

```python
env = "dev"
catalog_name = "capstone_src"

bronze_schema = f"{env}_bronze"
silver_schema = f"{env}_silver"
gold_schema = f"{env}_gold"

users_table_name = "users"
products_table_name = "products"
sales_table_name = "sales"

users_gold = "dim_user"
products_gold = "dim_product"
sales_gold = "fact_sales"

full_users_bronze = f"{catalog_name}.{bronze_schema}.{users_table_name}"
full_users_silver = f"{catalog_name}.{silver_schema}.{users_table_name}"
full_users_gold = f"{catalog_name}.{gold_schema}.{users_gold}"

full_products_bronze = f"{catalog_name}.{bronze_schema}.{products_table_name}"
full_products_silver = f"{catalog_name}.{silver_schema}.{products_table_name}"
full_products_gold = f"{catalog_name}.{gold_schema}.{products_gold}"

full_sales_bronze = f"{catalog_name}.{bronze_schema}.{sales_table_name}"
full_sales_silver = f"{catalog_name}.{silver_schema}.{sales_table_name}"
full_sales_gold = f"{catalog_name}.{gold_schema}.{sales_gold}"
```

## Batch Users

```python
display(spark.sql(f"SELECT * FROM {full_users_bronze} ORDER BY user_id, last_modified"))
display(spark.sql(f"SELECT * FROM {full_users_silver} ORDER BY user_id"))
display(spark.sql(f"DESCRIBE HISTORY {full_users_bronze}"))
display(spark.sql(f"DESCRIBE HISTORY {full_users_silver}"))
```

## Batch Products

```python
display(spark.sql(f"SELECT * FROM {full_products_bronze} ORDER BY product_id, last_modified"))
display(spark.sql(f"SELECT * FROM {full_products_silver} ORDER BY product_id"))
display(spark.sql(f"DESCRIBE HISTORY {full_products_bronze}"))
display(spark.sql(f"DESCRIBE HISTORY {full_products_silver}"))
```

## CDC Merge

```python
display(spark.sql(f"SELECT * FROM {full_sales_bronze} ORDER BY sale_id, updated_at"))
display(spark.sql(f"SELECT * FROM {full_sales_silver} ORDER BY sale_id"))
display(spark.sql(f"DESCRIBE HISTORY {full_sales_bronze}"))
display(spark.sql(f"DESCRIBE HISTORY {full_sales_silver}"))
```

```python
display(spark.sql(f"SELECT * FROM {full_users_gold} ORDER BY user_id"))
display(spark.sql(f"SELECT * FROM {full_products_gold} ORDER BY product_id"))
display(spark.sql(f"SELECT * FROM {full_sales_gold} ORDER BY sale_date, product_id, user_id"))
```
